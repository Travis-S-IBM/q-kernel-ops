## Doc

The download function is only here if you don't alrealdy have the data files. To use it, had your github token in the var `token` and uncomment it.

To use this notebook, you don't need to download each data files each times. Comment the call to the download function when not need.

```python
def dl_feather(f_path, f_name):
    file = f_name
    f_path += file
    
    # define parameters for a request
    #token = '' 
    owner = 'Travis-S-IBM'
    repo = 'q-kernel-ops'

    # send a request
    r = requests.get(
        'https://api.github.com/repos/{owner}/{repo}/contents/{path}'.format(
        owner=owner, repo=repo, path=f_path),
        headers={
            'accept': 'application/vnd.github.v3.raw',
            'authorization': 'token {}'.format(token)
                }
        )
    
    open("../" + f_path, 'wb').write(r.content)
```

## Telemetry

```python
dl_feather("resources/kernel_metadata/", "telemetry_info.csv")
```
